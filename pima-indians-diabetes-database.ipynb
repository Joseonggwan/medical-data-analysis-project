```python
# Create a pipeline that standardizes the data then creates a model
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
X = diabetes.loc[:,diabetes.columns!="Outcome"]
y = diabetes.loc[:,'Outcome']
# create pipeline
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('lda', LinearDiscriminantAnalysis()))
model = Pipeline(estimators)
# evaluate pipeline
seed = 10
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(model, X, y, cv=kfold)
print(results.mean())
```

KFold은 데이터를 여러 번 나눠서 평가하는 것
cross_val_score은 모델 성능을 반복 측정
한 번만 평가하면 운에 따라 결과가 달라질 수 있어서 여러 번 나눠서 신뢰도 높은 성능을 평가합니다.
StandardScaler은 변수들의 단위를 맞추는 것(표준화)
혈압,나이,BMI 등 스케일이 각각 달라서 이대로 사용한다면 특정 변수의 영향이 커지게 됩니다. 변수들의 단위를 맞춤으로써 공정하게 모델을 학습시킵니다.
LDA 모델을 사용해서 분류 분석을 수행합니다. 입력 변수들(X)를 이용해서 Outcome(당뇨 여부)를 예측합니다.
X: 설명 변수(나이, 혈당 등)
Y: 목표 변수(당뇨 여부)
X를 통해 Y를 예측
파이프라인(전처리 + 모델)을 구성
데이터 누수를 방지한다.
랜덤 시드를 설정 seed = 10
결과를 재현 가능하게 만드는 것
데이터를 10개로 나눠서 반복적으로 학슴과 검증을 수행합니다.
모델이 특정 데이터에만 잘 맞는지 확인합니다.
모델 성능을 10번 반복 측정 -> 10개의 정확도 값 (모델이 얼마나 안정적인지 확인)
10개의 정확도 값을 평균을 내보니 약 0.77
새로운 데이터에서도 약 77% 정확도를 기대할 수 있습니다.
0.7~0.8은 보통이거나 괜찮은 성능을 가진다고 말할 수 있습니다.

이 데이터분석의 의의는 당뇨병 예측 모델이 실제 데이터에서도 안정적으로 작동하는지 검증한 분석이라고 볼 수 있습니다.
